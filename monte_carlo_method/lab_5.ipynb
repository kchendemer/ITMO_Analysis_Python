### Задача 2. <br>
Разделить таблицу данных на четыре одинаковых части.<br>

````python
#  Чтобы точнее разделить таблицу на четыре одинаковые части используем квартили
r1 = np.random.uniform(low = 0.0, high=1.0, size=len(df))

q1 = np.quantile(r1, 0.25)
q2 = np.quantile(r1, 0.5)
q3 = np.quantile(r1, 0.75)

q1, q2, q3
````

````python
df_1 = df.iloc[r1<q1]
df_2 = df.iloc[(r1>q1) & (r1<q2)]
df_3 = df.iloc[(r1>q2) & (r1<q3)]
df_4 = df.iloc[r1>q3]
````

### Задача 3 <br>
Разделить таблицу данных на две стратифицированные подвыборки. По переменным кока-кола и минералка.

````python
# Создаем объединенную колонку по переменным COKE и TAB
df['strat_coke_tab'] = df['COKE'].astype(str) + '_' + df['TAB'].astype(str)
df.head()
````

````python
# Проверяем распределение данных
df.strat_coke_tab.value_counts()
````

Так как присутствует группа с 1 значением, то при автоматическом (используя sklearn) делении на равные выборки выходит ошибка: <br>
```train_df, test_df = train_test_split(df, test_size = 0.5, stratify=df['strat_coke_tab'], shuffle=True, random_state=42)```

> ValueError: The least populated class in y has only 1 member, which is too few. The minimum number of groups for any class cannot be less than 2.

В таком случае будем делить вручную и при 1 значении отправлять в обучающую выборку.

````python
train_strat = []
test_strat = []

for i in df['strat_coke_tab'].unique():
    if len(df[df['strat_coke_tab'] == i].index) == 1:
        train_strat.append(df.loc[df[df['strat_coke_tab'] == i].index])
        continue

    group_test = int(len(df[df['strat_coke_tab'] == i].index) * 0.5)
    indx = np.random.permutation(df[df['strat_coke_tab'] == i].index)
    train_strat.append(df.loc[indx[:group_test]])
    test_strat.append(df.loc[indx[group_test:]])

train_strat_df = pd.concat(train_strat)
test_strat_df = pd.concat(test_strat)
````

````python
train_strat_df.head()
````

````python
test_strat_df.head()
````

````python
# Убедимся, что единственное значение попало в обучающую выборку
train_strat_df['strat_coke_tab'].value_counts()
````

````python
# Убедимся, что единственное значение не попало в тестовую выборку
test_strat_df['strat_coke_tab'].value_counts()
````

````python
# Удаляем столбец со стратифицированными данными
train_strat_df.drop(columns=['strat_coke_tab'], inplace=True)
test_strat_df.drop(columns=['strat_coke_tab'], inplace=True)
````

````python
# Проверяем соотношение выборок
print(f'Размер обучающей выборки: {len(train_strat_df.index)}, размер тестовой выборки: {len(test_strat_df.index)}')
print(f'Соотношение выборок: {np.round(len(train_strat_df.index)/len(df), 2)}, {np.round(len(test_strat_df.index)/len(df), 2)}')
````
